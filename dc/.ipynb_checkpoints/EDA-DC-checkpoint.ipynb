```python
grouped_traps = df.groupby(['Date','Trap','Species'])[['NumMosquitos', 'WnvPresent']].sum()
```

```python
flat_groups = grouped_traps.reset_index().set_index('Date')
```

```python
flat_groups['WnvPresent'] = flat_groups['WnvPresent'].map(lambda x: int(x > 0))
```

```python
train_data = pd.read_csv('../data/train.csv')
```

```python
train_data.columns
```

```python
flat_groups
train_data
```

```python
trap_names = train_data['Trap'].value_counts().index
```

```python
long, lat = {}, {}
for trap in trap_names:
    mask = train_data['Trap'] == trap
    filt_data = train_data[mask]
    trap_lon = filt_data.iloc[0, 8]
    trap_lat = filt_data.iloc[0, 7]
    long[trap] = trap_lon
    lat[trap] = trap_lat
```

```python
flat_groups.head()
```

```python
flat_groups['longitude'] = flat_groups.Trap.map(long)
flat_groups['latitude'] = flat_groups.Trap.map(lat)
```
