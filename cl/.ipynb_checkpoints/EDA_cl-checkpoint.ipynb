```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
df = pd.read_csv('../data/final_dataframe.csv', index_col = 0)
```

```python
df.columns
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x=df['Species'], y=df['NumMosquitos'], hue = df['WnvPresent'], palette = 'rainbow_r', ci=None)
plt.xlabel('Species', size=25)
plt.xticks(size=13)
plt.ylabel('Number of Mosquitos', size=25)
plt.yticks(size=23)
plt.legend(labels=['Not Present','Present'], fontsize=20);
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(x=df['Month'], y=df['Tmax'], hue = df['WnvPresent'], palette = 'rainbow_r', ci=None, lw=7)
plt.xlabel('Month', size=35)
plt.xticks()
plt.ylabel('Max Temperature', size=35)
plt.yticks(size=23)
plt.legend(labels=['WNV Not Present','WNV Present'], fontsize=20);
```
